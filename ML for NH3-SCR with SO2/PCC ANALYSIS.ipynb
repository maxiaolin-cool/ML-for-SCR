```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# 1. Excel
df = pd.read_excel(' ().xlsx')  # 

# 2. PCC
corr_matrix = df.corr(method='pearson')

# 3. --
colors = ["#9BB7D4", "#FFFFFF", "#F5C3C2"]  #  ->  -> 
cmap = LinearSegmentedColormap.from_list("blue_white_pink", colors)

# 4. 
sns.set_style("whitegrid")  # +
plt.rcParams.update({
    'font.family': 'Arial',      # 
    'font.weight': 'bold',       # 
    'axes.labelweight': 'bold',  # 
    'axes.titleweight': 'bold'   # 
})

# 5.  - 
plt.figure(figsize=(6.8, 3), dpi=300)  # 

heatmap = sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",        # 
    cmap=cmap,
    center=0,         # 0
    vmin=-1,          # -1
    vmax=1,           # 1
    linewidths=0.5,   # 
    linecolor='#F0F0F0',  # 
    cbar_kws={"shrink": 0.8, "label": "Correlation", "orientation": "horizontal"}  # 
)

# 6. 
plt.title('', fontsize=16, pad=20)

#  - x0
plt.xticks(fontsize=7, rotation=0, ha='center')  # x
plt.yticks(fontsize=7, rotation=0)  # y

# 
for text in heatmap.texts:
    text.set_size(5)  # 
    text.set_weight('bold')  # 

# 
cbar = heatmap.collections[0].colorbar
cbar.ax.tick_params(labelsize=7)  # 
cbar.ax.set_xlabel('Correlation', fontsize=8, weight='bold')  # 
cbar.ax.xaxis.set_label_position('top')  # 

# 
cbar.ax.set_position([0.2, 0.05, 0.6, 0.03])  # [left, bottom, width, height]

# 7. 
plt.tight_layout()
plt.savefig('pcc_heatmap_blue_white_pink.png', 
            bbox_inches='tight', 
            dpi=300,
            facecolor='white')  # 
plt.show()
```
